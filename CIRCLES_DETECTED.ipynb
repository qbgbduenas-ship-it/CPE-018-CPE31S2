Write a Python script using OpenCV that:

Load and preprocess the image
Detect circles using Hough Circle Transform
Draw detected circles in red color
Save the result as circles_detected.png

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/drive/MyDrive/EMTECH1/circles.png")
if img is None:
    raise FileNotFoundError("ERROR!!! : ((() ))")

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.GaussianBlur(gray, (9, 9), 2)

circles = cv2.HoughCircles(gray,cv2.HOUGH_GRADIENT,param1=180,dp=1.1,minRadius=30,maxRadius=250,minDist=110,param2=43)

if circles is not None:
    circles = np.uint16(np.around(circles))
    print("CIRCLESSSSSS", len(circles[0]))

    for (x, y, r) in circles[0]:
        cv2.circle(img, (x, y), r, (0, 0, 255), 4)
        cv2.circle(img, (x, y), 2, (0, 0, 255), 3)

output_image_path = "/content/drive/MyDrive/EMTECH1/circles.png"
cv2.imwrite("circles_detected.png", img)
cv2_imshow(img)
print("Saved as circles_detected.png")

```
