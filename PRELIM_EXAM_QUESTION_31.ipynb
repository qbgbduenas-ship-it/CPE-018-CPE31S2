Write a Python script using OpenCV that:

Reads an image named "nature.jpg"
Converts it to grayscale
Resizes it to 300x300 pixels
Saves the result as "processed.jpg"


```python
import cv2
from google.colab import drive
import os

drive.mount('/content/drive')
image_path = "/content/drive/MyDrive/EMTECH/nature.jpg"


image = cv2.imread(image_path)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
resized_image = cv2.resize(gray_image, (300, 300))

print("Image has been processed and saved to Google Drive as 'processed.jpg'")
from google.colab.patches import cv2_imshow

cv2_imshow(resized_image)
```
